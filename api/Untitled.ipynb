```python
#импорт библиотек
import pandas as pd
import numpy as np
import os
import fnmatch
import warnings
import json
```

```python
directory = "./"
files = []
files += os.listdir(directory)

file_list = []
for file in files:
    if fnmatch.fnmatch(file, '2023-07-*.xlsx'):
        file_list.append(file)

```

```python
#объединение таблиц 
all_tables = pd.read_excel(file_list[0], sheet_name='Sheet1')
for i in range(1, len(file_list)):
    df = pd.read_excel(file_list[i], sheet_name='Sheet1')
    all_tables = pd.concat([all_tables, df], axis=0)

all_tables = all_tables.drop("lot_id",axis=1)
all_tables = all_tables.drop("winner_ship_id",axis=1)
all_tables
```

```python
#таблицы с дополнительными данными
docks = pd.read_excel('docks.xlsx', sheet_name='Sheet1')[['id', 'name']]
routes = pd.read_excel('routes.xlsx', sheet_name='Sheet1')[['id', 'name']]
shipowners = pd.read_excel('shipowners.xlsx', sheet_name='Sheet1')[['id', 'name']]
ships = pd.read_excel('ships.xlsx', sheet_name='Sheet1')[['id', 'name']]
```

```python
#связывание объединенных таблиц с доплнительными данными
all_with_merge = pd.merge(all_tables, ships,how='left',left_on='ship_id',right_on='id', suffixes=('', '_sh')).drop("id_sh",axis=1)
all_with_merge = pd.merge(all_with_merge, docks,how='left',left_on='dock_id',right_on='id', suffixes=('', '_dock')).drop("id_dock",axis=1)
all_with_merge = pd.merge(all_with_merge, ships,how='left',left_on='ship_id',right_on='id', suffixes=('', '_ship')).drop("id_ship",axis=1)
all_with_merge = pd.merge(all_with_merge, shipowners,how='left',left_on='shipowner_id',right_on='id', suffixes=('', '_shipowner')).drop("id_shipowner",axis=1)
all_with_merge = pd.merge(all_with_merge, routes,how='left',left_on='route_id',right_on='id', suffixes=('', '_route')).drop("id_route",axis=1)
all_with_merge.drop("name",axis=1)
```

```python
#сохранение итоговой таблицы в xlsx формате
all_with_merge.to_excel('Done_all.xlsx',sheet_name='First_Sheet',index=False)
```

```python
#сохранение итоговой таблицы в json формате
json_str = all_with_merge.to_json(orient='records', force_ascii=False)

with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(json_str, f, ensure_ascii=False, indent=4)
```
